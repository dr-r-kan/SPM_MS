# Python Microstate Comparison Analysis

Notebook wrapper for `analyze_comparison_results.py`. The script is the source of truth; this notebook just selects paths, runs it, and displays the main figures.

```python
from pathlib import Path
from IPython.display import Image, display

import analyze_comparison_results as acr

candidates = [Path('outputs/simulations/results/comparison_results.csv')]
candidates += sorted(Path('outputs').glob('*/results/comparison_results.csv'))
RESULTS_CSV = next((p for p in candidates if p.is_file()), None)
if RESULTS_CSV is None:
    raise FileNotFoundError('No comparison_results.csv found under outputs/. Run the MATLAB simulation first or set RESULTS_CSV manually.')

RESULTS_DIR = RESULTS_CSV.parent
OUTPUT_DIR = RESULTS_DIR.parent / 'analysis_plots_python_notebook'
TEMPLATE_FILE = Path('MetaMaps_2023_06.set')

RESULTS_CSV, OUTPUT_DIR
```

```python
result = acr.analyze_comparison_results(
    RESULTS_DIR,
    results_csv=RESULTS_CSV,
    output_dir=OUTPUT_DIR,
    template_file=TEMPLATE_FILE,
    n_boot=200,
    n_folds=2,
    seed=0,
    max_confusion_pair_plots=0,
    run_mixed_models=True,
)

print('Statistics:', result['stats_file'])
print('Plots:', result['plots_dir'])
result['table'].head()
```

```python
main_figures = [
    'boxplot_comparison.png',
    'method_effects_with_ci.png',
    'criterion_effects_with_ci.png',
    'snr_effects.png',
    'interaction_plot.png',
    'avg_k_error_heatmap.png',
    'k_estimated_vs_true_heatmap.png',
    'abs_k_error_by_method.png',
    'runtime_snr_effect.png',
    'simulated_microstate_method_comparison_seed0.png',
    'backfit_confusion_summary.png',
]

for name in main_figures:
    path = OUTPUT_DIR / name
    if path.is_file():
        print(path)
        display(Image(filename=str(path)))
```

```python
comparison_dir = OUTPUT_DIR / 'backfit_confusion_comparisons'
manifest = comparison_dir / 'backfit_confusion_comparison_manifest.csv'
if manifest.is_file():
    print(manifest)
    display(acr.pd.read_csv(manifest).head())

for path in sorted(comparison_dir.glob('*.png'))[:3]:
    print(path)
    display(Image(filename=str(path)))
```
